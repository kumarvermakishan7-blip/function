1. What is the difference between a function and a method in Python?
- **Function**
- A function is a block of reusable code that performs a specific task.
- Defined using def keyword, independently of any class.
- Called directly using its name.
- Functions exist independently.
- **Method**
- A method is a function that is associated with an object or class.
- Defined inside a class and usually works on the data (attributes) of that class.
- Called on an object using dot notation.
- Methods always belong to an object/class.



```python
# Function Example
def add_numbers(a, b):
    return a + b

print(add_numbers(5, 3))

```

```python
# Method Example
class Calculator:
    def multiply(self, x, y):
        return x * y

calc = Calculator()
print(calc.multiply(4, 6))
```

3. What are the different ways to define and call a function in Python?


- **Normal Function Definition**

- Defined using the def keyword.

- Called directly by its name.

```python

def greet():
    print("Hello, World!")

greet()
```

- **Function with Parameters**
- Accepts input values (parameters).

- Called with arguments.

```python
def add(a, b):
    return a + b

print(add(4, 6))
```

- **Function with Default Parameters**
- Parameters can have default values.

- If no argument is passed, default value is used.

```python
def greet(name="Guest"):
    print("Hello,", name)

greet("Kishan")
greet()
```

- **Function with Variable-Length Arguments**
- *args : accepts multiple positional arguments.

- **kwargs : accepts multiple keyword arguments.

```python
def show_args(*args):
    print(args)

show_args(1, 2, 3)


def show_kwargs(**kwargs):
    print(kwargs)

show_kwargs(name="Kishan", age=20)
```

- **Lambda (Anonymous) Functions**
- Functions defined in a single line using lambda.

- Called directly where needed.

```python
square = lambda x: x * x
print(square(5))
```

- **Recursive Function**
- A function that calls itself.

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

6. Explain the concept of generators in Python and how they are defined.
- A generator is a special type of function in Python that is used to produce a sequence of values one at a time instead of returning them all at once.

- They are memory-efficient because values are generated on demand.

- A generator uses the yield keyword instead of return.

- Each time the generator’s __next__() method is called (or next() function is used), execution resumes from where it was last paused.


```python
def count_up_to(n):
    for i in range(1, n+1):
        yield i

gen = count_up_to(3)
print(next(gen))  # 1
```

7. What are the advantages of using generators over regular functions?
- Memory Efficient → They generate values one by one instead of storing all values in memory.
Example: Generating numbers from 1 to 1,000,000 without storing them in a list.

- Lazy Evaluation → Values are produced only when needed, saving time and resources.
Example: Reading a large file line by line.

- Infinite Sequences → Can represent endless streams of data without running out of memory.
Example: Generating Fibonacci numbers.

- Cleaner Code → No need to write complex iterator classes; yield handles it automatically.


8. What is a lambda function in Python and when is it typically used?
- A lambda function is an anonymous (nameless) function defined using the lambda keyword.

- It can have any number of arguments but only one expression.

- Syntax:
 lambda arguments: expression
- **When Used:**
- When a small, one-time function is needed.
 - Commonly used with functions like map(), filter(), sorted(), or inside short calculations.

```python

def square(x):
    return x*x


square_lambda = lambda x: x*x

print(square_lambda(5))
```

```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)   # Output: [1, 4, 9, 16]
```

9. Explain the purpose and usage of the map() function in Python.,with example
- The map() function in Python applies a given function to each item of an iterable and returns a map object (iterator).
- map(function, iterable)
- It’s mainly used to avoid writing loops when applying the same operation to all elements.

```python
nums = [10, 20, 30, 40]
result = map(lambda x: x/10, nums)
print(list(result))   # [1.0, 2.0, 3.0, 4.0]

```

10. What is the difference between map(), reduce(), and filter() functions in Python?
- **map()**

- Purpose: Apply a function to every element in an iterable.

- Output: A new iterable (map object).

- **filter()**

- Purpose: Select (filter) elements from an iterable based on a condition (True/False).

- Output: A filtered iterable.

- **reduce() (from functools)**

- Prpose: Repeatedly apply a function to reduce the iterable to a single value.

- Output: One value.

```python
#map
nums = [1, 2, 3, 4]
print(list(map(lambda x: x**2, nums)))   # [1, 4, 9, 16]

```

```python
#filter
nums = [1, 2, 3, 4, 5, 6]
print(list(filter(lambda x: x % 2 == 0, nums)))   # [2, 4, 6]
```

```python
#reduce
from functools import reduce
nums = [1, 2, 3, 4]
print(reduce(lambda x, y: x + y, nums))   # 10

```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            val = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return val
        else:
            raise StopIteration




```

```python
for num in Fibonacci(5):
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def poweroftwo(n):
  for i in range(n+1):
    yield 2**i
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```
